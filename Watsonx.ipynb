```python
import pandas as pd
from datetime import datetime

# Load the CSV file
csv_file = '/content/drive/MyDrive/Watsonx/cleaned_sfc_data.csv'
df = pd.read_csv(csv_file)

# Convert 'Anniversary Date' from YYYYMMDD to YYYY-MM-DD format
df['Anniversary Date'] = pd.to_datetime(df['Anniversary Date'], format='%Y%m%d')

# Calculate the age
current_date = datetime.now()
df['Age'] = df['Anniversary Date'].apply(lambda x: current_date.year - x.year - ((current_date.month, current_date.day) < (x.month, x.day)))

# Save the updated CSV with the new 'Age' column
df.to_csv(csv_file, index=False)

print("The 'Age' column has been added successfully.")
```

Cleaning Data

```python
import pandas as pd
import random
from datetime import datetime
from dateutil import parser

# File paths
file_path = '/content/drive/MyDrive/Watsonx/10kdata.csv'
cleaned_file_path = '/content/drive/MyDrive/Watsonx/10kdata_cleaned.csv'

# Load the dataset
df = pd.read_csv(file_path)

# Values to replace nulls with
values = [
    "Sofala",
    "Nampula",
    "Maputo Provincia",
    "Zambezia",
    "Inhambane",
    "Maputo Cidade",
    "Tete",
    "Gaza",
    "Niassa"
]

# Replace null values in CELLID_MOST_USED_PROVINCE with random choices from the values list
df['CELLID_MOST_USED_PROVINCE'] = df['CELLID_MOST_USED_PROVINCE'].apply(
    lambda x: random.choice(values) if pd.isnull(x) else x
)

# Value to replace nulls in SUB_BIRTHDATE
birthdate_value = "15/07/1987"

# Replace null values in SUB_BIRTHDATE with birthdate_value
df['SUB_BIRTHDATE'] = df['SUB_BIRTHDATE'].apply(
    lambda x: birthdate_value if pd.isnull(x) else x
)

# Function to calculate age from birthdate
def calculate_age(birthdate, reference_date=datetime.now()):
    birthdate = parser.parse(birthdate)
    age = reference_date.year - birthdate.year - ((reference_date.month, reference_date.day) < (birthdate.month, birthdate.day))
    return age

# Create a new column 'AGE' based on 'SUB_BIRTHDATE'
df['AGE'] = df['SUB_BIRTHDATE'].apply(calculate_age)

# Drop the SUB_BIRTHDATE column
df = df.drop(columns=['SUB_BIRTHDATE'])

# Save the updated dataset to a new CSV file
df.to_csv(cleaned_file_path, index=False)

print(f"Null values replaced, AGE column added, SUB_BIRTHDATE column removed, and cleaned dataset saved successfully to {cleaned_file_path}.")
```

##Script for Extracting NETWORK_DAYS

```python
import pandas as pd
from datetime import datetime

# Define the reference date (format: day/month/year)
reference_date = datetime.strptime('30/07/2024', '%d/%m/%Y')

# Load the Excel file
file_path_excel = '/content/drive/MyDrive/Watsonx/10kdata.xlsx'  # Replace with your file path
df = pd.read_excel(file_path_excel)

# Ensure that ACTIVATION_DATE column is in datetime format
df['ACTIVATION_DATE'] = pd.to_datetime(df['ACTIVATION_DATE'], format='%d/%m/%Y')

# Calculate the difference in days and create a new column 'network_days'
df['network_days'] = (reference_date - df['ACTIVATION_DATE']).dt.days

# Save the updated dataframe back to Excel in the same path
df.to_excel(file_path_excel, index=False)
```
